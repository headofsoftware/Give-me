# MODEL SELECTON

```python
from sklearn.metrics import f1_score,confusion_matrix,mean_absolute_error
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRegressor
```

```python
from sklearn.model_selection import train_test_split
best_col = ["RevolvingUtilizationOfUnsecuredLines" ,"NumberOfTime30-59DaysPastDueNotWorse", "NumberOfTimes90DaysLate"]
#RevolvingUtilizationOfUnsecuredLines 59DaysPastDueNotWorse NumberOfTimes90DaysLate
# inputs = dataset.drop(columns=["SeriousDlqin2yrs","DebtRatio","MonthlyIncome","NumberOfDependents","age","NumberOfTime30-59DaysPastDueNotWorse"])
inputs = dataset.drop(columns=["SeriousDlqin2yrs"])
outputs= dataset["SeriousDlqin2yrs"]
X, x, Y, y = train_test_split(inputs, outputs,test_size = 0.2, random_state = 42)
```

```python
xgb = XGBClassifier(n_jobs=-1) 
# Use a grid over parameters of interest
param_grid = {
                  'n_estimators' :[100,150,200,250,300],
                  "learning_rate" : [0.001,0.01,0.0001,0.05, 0.10 ],
                  "gamma"            : [ 0.0, 0.1, 0.2 , 0.3 ],
                  "colsample_bytree" : [0.5,0.7],
                  'max_depth': [3,4,6,8]
              }
xgb_randomgrid = RandomizedSearchCV(xgb, param_distributions=param_grid,verbose=0)
xgb_randomgrid.fit(X,Y)
xgb_randomgrid.best_score_
```

```python
y_pred = xgb_randomgrid.predict(x)
y_pred2 = xgb_randomgrid.predict_proba(x)[:,1]
```

```python
# 0.68   LogisticRegression

lr_woe = LogisticRegression(max_iter=300, solver = 'liblinear')
lr_woe.fit(X, Y.ravel())
lr_woe.score(x,y)
# 0.77   RandomForestClassifier

rfc = RandomForestClassifier().fit(X,Y)
rfc.score(x,y)
# 0.80   XGBClassifier
xgc = XGBClassifier().fit(X,Y)
xgc_pred = xgc.predict(x)
xgc_pred2 = xgc.predict_proba(x)[:,0]
print("Score :",xgc.score(X,Y))
print("F1 Score:",f1_score(y,xgc_pred))
print("Confusion Matrix",confusion_matrix(y,xgc_pred))
mean_absolute_error(y,xgc_pred2)
```

```python
rfc = RandomForestClassifier(n_estimators=500 , max_depth=7 ).fit(X,Y)
print(rfc.score(x,y))
rfc_pred = rfc.predict_proba(x)[:,1]
rfc.feature_importances_
```

```python
from sklearn.metrics import auc,roc_curve
fpr,tpr,_ = roc_curve(y, xgr_pred)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(10,8))
plt.title('Receiver Operating Characteristic')
sns.lineplot(fpr, tpr, label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
test2 = pd.read_csv("../input/GiveMeSomeCredit/cs-test.csv")
id_col = test2["Unnamed: 0"]
test = test.drop(columns=["SeriousDlqin2yrs"])
pred = rfc.predict_proba(test)[:,1]
col = [pred,id_col.values]
result = pd.DataFrame(list(zip(id_col, pred)),columns=["Id","Probability"])
result.to_csv("result.csv",index=False)
```
